```python
def bucketize(review_list, review_key):
    return_bucket = {}
    for review in review_list:
        for item in review._fields:
            return_bucket.setdefault(item, []).append(getattr(review,item))
    return return_bucket[review_key]
```

```python
import pandas as pd

def scatter(x, y, xlabel="x axis", ylabel="y axis"):
    df = pd.DataFrame({"x":x, "y":y})
    ax = df.plot.scatter(x="x", y="y", color="black", fontsize=16, xlim=0, ylim=0)
    ax.set_xlabel(xlabel, fontsize=16)
    ax.set_ylabel(ylabel, fontsize=16)
    ax.get_xaxis().get_major_formatter().set_scientific(False)
    ax.get_yaxis().get_major_formatter().set_scientific(False)
```

```python
def read_json2(file):
    f = open(file, encoding = "utf-8")
    json_read = f.read()
    f.close()
    data = json.loads(json_read)
    return data
```

```python
def combine_json(list_of_paths):
    curdir = {}
    masterdir = {}
    for path in list_of_paths:
        curdir[path] = read_json2(path)
    for key in curdir:
        for otherkey in curdir[key]:
            masterdir[otherkey] = curdir[key][otherkey]
    return masterdir
```
